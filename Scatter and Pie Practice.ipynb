```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c="red", label="Chicago")

plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.gca().invert_yaxis()

plt.xlim(0, 50)
plt.title("PyBer Fare by Month")

plt.legend()
```

```python
plt.scatter(x_axis, y_axis, c="red", label="Chicago", s=y_axis)

plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.gca().invert_yaxis()

plt.ylim(0, 50)
plt.title("PyBer Fare by Month")

plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, c="red", label="Chicago", s=y_axis_larger)

plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.gca().invert_yaxis()

plt.ylim(0, 50)
plt.title("PyBer Fare by Month")

plt.legend()
```

```python
# Can do the same thing with a list comprehension
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
# Now using OOP
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*3 for i in y_axis], color='skyblue', alpha=0.8, linewidth=2, edgecolor="black", label="Boston")
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.title("PyBer Fares by Month in Boston")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
```

```python
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# subplots MUST be before the plot is set
plt.subplots(figsize=(8,8))
plt.pie(
    y_axis, 
    labels=x_axis, 
    explode=explode_values, 
    autopct = "%.1f%%", 
    colors=colors
)

plt.show()
```

```python
# Same thing with OOP 
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["crimson", 
          "tomato", 
          "orangered", 
          "orange", 
          "yellow", 
          "chartreuse", 
          "green", 
          "teal", 
          "blue", 
          "navy", 
          "indigo", 
          "purple"]

fig, ax = plt.subplots(figsize=(9,9))
ax.pie(y_axis, 
       labels=x_axis, 
       colors=colors, 
       shadow=True, 
       explode=explode_values,
       startangle=-90
      )
plt.gca().invert_yaxis()
plt.show()
```
